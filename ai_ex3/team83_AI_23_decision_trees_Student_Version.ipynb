Θα δοκιμάσουμε για features = ["budget" , "popularity" , "duration" , "vote_average" , "num_voted_users" , "title_year"] αλλά και για όλα τα genres τα οποία είναι genres = ["Action" , "Adventure" , "Fantasy" , "Science Fiction" , "Crime" , "Drama" , "Thriller" , "Family" , "Western" , "Comedy" , "Animation" ,

Πρώτα φτιάχνουμε dictionaries τα οποία θα μας βοηθήσουν στην κωδικοποίηση της συνάρτησης MultiLabelEncoder για τα attributes με πολλαπλές τιμές δηλαδή genres, plot_keywords, production_companies, production_countries, spoken_languages.

```python
# merge two datasets
movies_metadata["movieId"] = movies_metadata["id"]
dataset = pd.merge(movies_metadata, ratings, on='movieId')
dataset = dataset[dataset['rating'].notna()]

##### let's study our dataset a bit more
genres_dict = {}
count = 0
for row in dataset.itertuples(index=True, name='Pandas'):
        ge = getattr(row, 'genres')
        if type(ge) == str:
          for g in ge.split("|"):
            if g not in genres_dict:
              genres_dict[g] = count
              count += 1

print("So we have" , len(genres_dict) , "genres which are:")
print(genres_dict)
print()

keywords_dict = {}
count = 0
for row in dataset.itertuples(index=True, name='Pandas'):
        ge = getattr(row, 'plot_keywords')
        if type(ge) == str:
          for g in ge.split("|"):
            if g not in keywords_dict:
              keywords_dict[g] = count
              count += 1

print("So we have" , len(keywords_dict) , "keywords which are:")
print(keywords_dict)
print()

companies_dict = {}
count = 0
for row in dataset.itertuples(index=True, name='Pandas'):
        ge = getattr(row, 'production_companies')
        ge = ast.literal_eval(ge)
        for g in ge:
          g = g['id']
          if g not in companies_dict:
            companies_dict[g] = count
            count += 1

print("So we have" , len(companies_dict) , "production companies with IDs:")
print(companies_dict)
print()

countries_dict = {}
count = 0
for row in dataset.itertuples(index=True, name='Pandas'):
        ge = getattr(row, 'production_countries')
        ge = ast.literal_eval(ge)
        for g in ge:
          g = g['iso_3166_1']
          if g not in countries_dict:
            countries_dict[g] = count
            count += 1

print("So we have" , len(countries_dict) , "production countries with iso_3166_1:")
print(countries_dict)
print()

language_dict = {}
count = 0
for row in dataset.itertuples(index=True, name='Pandas'):
        ge = getattr(row, 'spoken_languages')
        ge = ast.literal_eval(ge)
        for g in ge:
          g = g['iso_639_1']
          if g not in language_dict:
            language_dict[g] = count
            count += 1

print("So we have" , len(language_dict) , "spoken languages with iso_639_1:")
print(language_dict)
print()

directors_dict = {}
count = 0
for row in dataset.itertuples(index=True, name='Pandas'):
        ge = getattr(row, 'director_name')
        if type(ge) == str:
          if ge not in directors_dict:
            directors_dict[ge] = count
            count += 1

print("So we have" , len(directors_dict) , "directors:")
print(directors_dict)
print()

actors_dict = {}
count = 0
for row in dataset.itertuples(index=True, name='Pandas'):
        a1 = getattr(row, 'actor_1_name')
        a2 = getattr(row, 'actor_2_name')
        a3 = getattr(row, 'actor_3_name')
        if type(a1) == str:
          if a1 not in actors_dict:
            actors_dict[a1] = count
            count += 1
        if type(a2) == str:
          if a2 not in actors_dict:
            actors_dict[a2] = count
            count += 1
        if type(a3) == str:
          if a3 not in actors_dict:
            actors_dict[a3] = count
            count += 1

print("So we have" , len(actors_dict) , "actors:")
print(actors_dict)
print()
```

```python
# preprocess values
def MultiLabelEncoder(dataset, column_name):
    """

    Η συνάρτηση αυτή μετατρέπει μια στήλη με πολλαπλές τιμές σε κάθε κελί σε ένα δίανυσμα. Τέτοιες στήλες στο σύνολό μας είναι
    η στήλη  'genres' και 'plot_keywords'.
    Αν για παράδειγμα το genres για μια ταινία είναι "Action"|"Horror" ενώ τα συνολικά genres είναι k
    τότε η συνάρτηση πρέπει να επιστρέφει ένα διάνυσμα k θέσεων με μηδενικά εκτός από τις θέσεις που αντιστοιχούν στο Action και το Horror,
    όπου θα είναι 1.
    Είσοδος:
        dataset: ένα dataframe
        column_name: το όνομα του κελίου το όποιο θέλουμε να μετατρέψουμε σε διάνυσμα

    'Έξοδος: ένα διάνυσμα
    """
    ##################
    ## Your code below
    res = []

    if column_name == "genres":
      for row in dataset.itertuples(index=True, name='Pandas'):
        a = [0 for _ in range(len(genres_dict))]
        ge = getattr(row, 'genres')
        if type(ge) == str:
          for g in ge.split("|"):
            a[genres_dict[g]] = 1
        res.append(a)


    elif column_name == "plot_keywords":
      for row in dataset.itertuples(index=True, name='Pandas'):
        a = [0 for _ in range(len(keywords_dict))]
        key = getattr(row, 'plot_keywords')
        if type(key) == str:
          for g in key.split("|"):
            a[keywords_dict[g]] = 1
        res.append(a)

    elif column_name == "production_companies":
      for row in dataset.itertuples(index=True, name='Pandas'):
        a = [0 for _ in range(len(companies_dict))]
        prod = getattr(row, 'production_companies')
        prod = ast.literal_eval(prod)
        for studio in prod:
          studio_id = studio['id']
          a[companies_dict[studio_id]] = 1
        res.append(a)


    elif column_name == "production_countries":
      for row in dataset.itertuples(index=True, name='Pandas'):
        a = [0 for _ in range(len(countries_dict))]
        prod = getattr(row, 'production_countries')
        prod = ast.literal_eval(prod)
        for country in prod:
          country = country['iso_3166_1']
          a[countries_dict[country]] = 1
        res.append(a)


    elif column_name == "spoken_languages":
      for row in dataset.itertuples(index=True, name='Pandas'):
        a = [0 for _ in range(len(language_dict))]
        lang = getattr(row, 'spoken_languages')
        lang = ast.literal_eval(lang)
        for la in lang:
          la = la['iso_639_1']
          a[language_dict[la]] = 1
        res.append(a)

    elif column_name == "budget":
      for row in dataset.itertuples(index=True, name='Pandas'):
        a = []
        budget = getattr(row, 'budget')
        budget = int(budget)
        if budget < 25000000:   ## 25 million just a number that makes sense according to 1.b
          a = [0]    ## our user seems to enjoy movies with a big budget
        else:
          a = [1]
        res.append(a)

    elif column_name == "duration":
      for row in dataset.itertuples(index=True, name='Pandas'):
        a = []
        duration = getattr(row, 'duration')
        duration = int(duration)
        if duration < 100:
          a = [0]   ## our user seems to enjoy longer movies
        else:
          a = [1]
        res.append(a)

    elif column_name == "title_year":
      for row in dataset.itertuples(index=True, name='Pandas'):
        a = []
        year = getattr(row, 'title_year')
        year = int(year)
        if year < 2006:
          a = [0]   ## our user seems to enjoy modern movies
        else:
          a = [1]
        res.append(a)

    elif column_name == "director_name":
      for row in dataset.itertuples(index=True, name='Pandas'):
        a = [0 for _ in range(len(directors_dict))]
        dir = getattr(row, 'director_name')
        if type(dir) == str:
          a[directors_dict[dir]] = 1
        res.append(a)

    elif column_name == "actor_name":
      for row in dataset.itertuples(index=True, name='Pandas'):
        a = [0 for _ in range(len(actors_dict))]
        a1 = getattr(row, 'actor_1_name')
        a2 = getattr(row, 'actor_2_name')
        a3 = getattr(row, 'actor_3_name')
        if type(a1) == str:
          a[actors_dict[a1]] = 1
        if type(a2) == str:
          a[actors_dict[a2]] = 1
        if type(a3) == str:
          a[actors_dict[a3]] = 1
        res.append(a)




    ## Your code above
    ##################
    return  res # το δίανυσμα που κωδικοποιεί την τιμή του κελίου
```

Για τώρα ας δούμε τα αποτελέσματα με genres και plot_keywords:

```python
import numpy as np

class Node:
    def __init__(self, feature_index=None, threshold=None, value=None, left=None, right=None):
        self.feature_index = feature_index  # Index of the feature to split on
        self.threshold = threshold  # Threshold value for the feature
        self.value = value  # Predicted value at the node (for leaf nodes)
        self.left = left  # Left child node
        self.right = right  # Right child node

class CARTClassifier:
    def __init__(self, max_depth=None, min_samples_split=2):
        self.max_depth = max_depth  # Maximum depth of the tree
        self.min_samples_split = min_samples_split  # Minimum number of samples required to split a node

    def fit(self, X, y):
        self.num_classes = len(np.unique(y))
        self.tree = self._build_tree(X, y)

    def predict(self, X):
        return [self._predict_sample(x, self.tree) for x in X]

    def _build_tree(self, X, y, depth=0):
        num_samples, num_features = X.shape
        ##################
        ## Your code below
        gini = self._gini_index(y)
        num_samples_per_class = np.bincount(y)
        ## Your code above
        ##################

        if depth == self.max_depth or num_samples < self.min_samples_split or gini == 0:
            leaf_value = np.argmax(num_samples_per_class)
            return Node(value=leaf_value)

        best_gini = float('inf')
        best_feature_index = None
        best_threshold = None

        for feature_index in range(num_features):
            thresholds = np.unique(X[:, feature_index])
            for threshold in thresholds:
                left_indices = X[:, feature_index] <= threshold
                right_indices = X[:, feature_index] > threshold

                left_gini = self._gini_index(y[left_indices])##
                right_gini = self._gini_index(y[right_indices])##
                weighted_gini = (left_gini * np.sum(left_indices) + right_gini * np.sum(right_indices)) / num_samples    ###
                if weighted_gini < best_gini:
                    best_gini = weighted_gini
                    best_feature_index = feature_index
                    best_threshold = threshold

        left_indices = X[:, best_feature_index] <= best_threshold
        right_indices = X[:, best_feature_index] > best_threshold

        left_child = self._build_tree(X[left_indices], y[left_indices], depth + 1)
        right_child = self._build_tree(X[right_indices], y[right_indices], depth + 1)

        return Node(feature_index=best_feature_index, threshold=best_threshold, left=left_child, right=right_child)

    def _gini_index(self, y):
        ##################
        ## Your code below
        if len(y) == 0:
          return 0

        zero_cat = [x for x in y if x == 0]
        one_cat = [x for x in y if x == 1]

        class_counts = np.bincount(y)
        class_probabilities = class_counts / len(y)

        if len(zero_cat) == 0:
          left_gini = 0
        else:
          left_gini = 1.0 - np.sum(class_probabilities**2)

        if len(one_cat) == 0:
          right_gini = 0
        else:
          right_gini = 1.0 - np.sum(class_probabilities**2)

        left_weight = len(zero_cat) / len(y)
        right_weight = len(one_cat) / len(y)

        gini_index = left_weight * left_gini + right_weight * right_gini

        ## Your code above
        ##################
        return gini_index # return a single value

    def _predict_sample(self, x, node):
        if node.value is not None:
            return node.value

        if x[node.feature_index] <= node.threshold:
            return self._predict_sample(x, node.left)
        else:
            return self._predict_sample(x, node.right)
```

Όσον αφορά ποιες άλλες μετρικές θα μπορούσαμε να χρησιμοποιήσουμε για να αξιολογήσουμε τον recommender μας, θα μπορούσαμε (όπως είχαμε και στην 2η εργαστηριακή άσκηση) να ελέγξουμε τα **precision, recall και f1** ([accuracy, precision, recall ,f1 ](https://towardsdatascience.com/accuracy-precision-recall-or-f1-331fb37c5cb9)). Το πρώτο θα έλεγχε τον αριθμό των False Positives δηλαδή πόσες ταινίες από αυτές που προτείνουμε δεν θα αρέσουν πραγματικά στο χρήστη ενώ το δεύτερο τα False Negatives δηλαδή πόσες ταινίες από αυτές που απορρίψαμε θα άρεσαν στον χρήστη. Το f1 θα υπολόγιζε τον αρμονικό μέσο (harmonic mean) των precision και recall.

Επίσης θα μπορούσαμε να χρησιμοποιούσαμε το **Normalized Discounted Cumulative Gain** ([NDCG](https://arize.com/blog-course/ndcg/)) που χρησιμοποιείται για αξιολόγηση της απόδοσης μηχανών αναζήτησης και recommendation systems.

Τέλος, δύο άλλα metrics που θα μπορούσαμε να χρησιμοποιήσουμε είναι τα **Novelty and Diversity** ([Novelty and Diversity](https://citeseerx.ist.psu.edu/document?repid=rep1&type=pdf&doi=ecb23352b3fd8abd32332790fda7aca59c498fdf)) ώστε να ελέγχουμε εάν οι ταινίες που προτείνουμε στο χρήστη του είναι γνώριμες. Αυτό θα βοηθήσει καθώς οι χρήστες σήμερα ζητάνε συνεχώς νέο και ενδιαφέρον περιεχόμενο.
