Σε αυτό το σημείο θα γίνει μια αναβάθμιση του συστήματος έτσι ώστε να παράγονται καλύτερες  συστάσεις οι όποιες θα λαμβάνουν υπόψιν και τις προτιμήσεις του χρήστη, οι όποιες θα εξάγονται από τις αξιολογήσεις που έχει κάνει μέχρι στιγμής. 

Η εκπαίδευση του recommender θα γίνεται ως εξής:

Για κάθε ταινία θα υπάρχει ένα score το όποιο αρχικά θε είναι ίσο με 0 και θα διαμορφώνεται από τις αξιολογήσεις κάθε user. Έτσι για έναν χρήστη με βάση τις αξιολογήσεις που υπάρχουν στο αρχείο train_ratings θα πρέπει:

1.	Για κάθε ταινία που έχει βαθμολογήσει να βρίσκονται οι κοινές ταινίες ανά κλίμακα και στο μέχρι τώρα σκορ κάθε παρόμοιας ταινίας θα προστίθεται ένα βάρος το όποιο θα μπορούσε να είναι το ποσοστό ομοιότητας της ταινίας (δηλαδή ένα βάρος για κάθε κλίμακα, αν δύο ταινίες μοιάζουν ενισχύουμε το βάρος που προσθέτουμε από το να μοιάζουν λιγότερο) επί τον βαθμό που έχει βάλει ο χρήστης για την αρχική ταινία (διαφορετικό είναι ο χρήστης να έχει βάλει 5/5 ή 1/5 σε μια ταινία από 3/5). 

2. Στη συνέχεια, ανάλογα με το σκορ που έχει σχηματιστεί για κάθε ταινία, θα επιλέγεται αν αυτή θα μπορούσε να είναι προτεινόμενη για τον χρήστη ή όχι και θα μετράμε το πόσο καλά τα πήγε το σύστήμά μας με βάση ορισμένες μετρικές.

Η λογική πίσω από την παραπάνω διαδικασία είναι ότι παρόμοιες ταινίες θα έχουν ανάλογο βαθμό. Για παράδειγμα αν ένας χρήστης έχει αξιολογήσει αρκετές  ταινίες οι όποιες είναι sci-fiction με 5/5 τότε μια ταίνια sci-fiction την όποια δεν έχει δεί λογικά θα του αρέσει και θα έπρεπε να την προτείνουμε.

Μετά την εκπαίδευση του συστήματος σας καλείστε να δοκιμάσετε τον recommender που κατασκευάσατε στην πράξη. Για τον σκοπό αυτό θα φορτώσετε το αρχείο test_ratings.csv όπου περιέχονται οι αξιολογήσεις του ίδιου χρήστη για άλλες ταινίες. Το σύστημά σας πρέπει να προβλέπει αν μια ταινία θα πρέπει να προταθεί στον χρήστη. Μια ταίνια θεωρούμε οτι προτείνεται στον  χρήστη, αν έχει βαθμό μεγαλύτερο του 3 στο test_ratings. Συνεπώς για την επίβλεψη του συστήματός σας θα πρέπει για κάθε μια από τις ταινίες του αρχείου test_ratings.csv να προβλέψετε εάν θα αρέσει στο χρήστη ή όχι ώστε να την προτείνετε. 

ΠΡΟΣΟΧΗ! Η "βαθμολογία" που θα υπολογίσετε για κάθε ταινία κατά το training δεν είναι απαραίτητα πρόβλεψη της βαθμολογίας που θα έβαζε ο χρήστης. 

Στην συνέχεια, σε συνδιασμό με τις πραγματικές απαντήσεις του χρήστη θα αξιολογήσετε το σύστημά σας χρησιμοποιόντας τις μετρικές: precision, recall, f1 οι οποίες είναι οι πλέον γνωστές μετρικές και ευρέως χρησιμοποιούμενες τεχνικές για την επίβλεψη-μέτρηση απόδοσης ανάλογων συστημάτων.

1. Precision: Δείχνει πόσο ακριβές είναι το σύστημα. Υπολογίζει πόσα από τα στιγμιότυπα τα όποια προβλέψαμε ότι ανήκουν σε μια κλάση όντως ανήκουν σε αυτή. Η μετρική αυτή μας δίνει μια εικόνα σχετικά με τον αριθμό των ταινίων που προβλέψαμε ως προτεινόμενες ενώ δεν θα έπρεπε.

2. Recall: Υπολογίζει πόσα από τα στιγμιότυπα που ανήκουν σε μια κλάση (π.χ. προτεινόμενες ταινίες) προβλέφθηκαν σωστά.

3. F1: Είναι ένας μέσος μεταξύ των παραπάνω δυο μετρικών, έτσι ώστε να διατηρείται μια ισορροπία μεταξύ τους. Υπολογίζεται από την παρακάτω σχέση:

$$F_1=2\frac{Precision\times{Recall}}{Precision+Recall}$$

Οι παραπάνω συναρτήσεις παρέχονται από την βιβλιοθήκη scikit-learn.

````python
import pandas as pd
from pyswip import Prolog
import ast        #### added this to remove double quotes when needed
````

````python
# Η βιβλιοθήκη pandas είναι χρήσιμη για την εργασία με τέτοια δεδομένα
import pandas as pd 
path = '/content/drive/MyDrive/ex2/'
# Διάβασμα του αρχείου 'movie_metadata.csv' 
data = pd.read_csv(path + "movies_metadata.csv") 
#Στο csv υπαρχούν κελία με nan τιμές
#Στις θέσεις αυτές βάζουμε 'UNK' πράγμα που  κάνουμε με την παρακάτω συνάρτηση
data.fillna("UNK", inplace=True)
# Preview the first 5 lines of the loaded data 
data.head()
````

````python
def clean_text(text):
  text = text.replace(u'\xa0', u'')
  text = text.replace(u"'", u'')
  return text
````

````python
#create World
#Ορίζουμε τον κόσμο μας
prolog = Prolog()

#Για κάθε row του πίνακα φτιάχνουμε τα κατηγορήματα που θέλουμε να αποθηκέυσουμε
#αρχικά σε μια λίστα με το όνομα literals
literals = []
movie_score = {}
for row in data.itertuples(index=True, name='Pandas'):
  movie_title = clean_text(getattr(row, 'movie_title'))
  
  for genre in getattr(row, 'genres').split("|"):
    if genre != "UNK": ## added this
      literals.append("genre('"+ movie_title +"','"+ genre +"')")      #### found the genres for queries 1 , 2 and 3
  
  ### YOUR CODE HERE \/

  director = clean_text(getattr(row, 'director_name'))
  if director != "UNK":
    literals.append("director('"+ movie_title +"','"+ director +"')")   #### found the director for query 4

  for keyword in getattr(row, 'plot_keywords').split("|"):
    keyword = clean_text(keyword)
    if keyword != "UNK":
      literals.append("plot_keyword('"+ movie_title +"','"+ keyword +"')")      #### found the plot keywords for queries 5 and 6 
  if 'black and white' in getattr(row,'plot_keywords').split('|'):
      literals.append(f"color('{movie_title}','black_n_white')")
  else:
      literals.append(f"color('{movie_title}','colored')")

  
  ### for the actors I don't think we need to make 3 seperate literals
  ### actor_1 , actor_2 and actor_3 but just one actor(Movie,Actor) but will see later
  
  actor1 = clean_text(getattr(row, 'actor_1_name'))
  actor2 = clean_text(getattr(row, 'actor_2_name'))
  actor3 = clean_text(getattr(row, 'actor_3_name'))
  if actor1 != "UNK":
    literals.append("actor('"+ movie_title +"','"+ actor1 +"')")
  if actor2 != "UNK":
    literals.append("actor('"+ movie_title +"','"+ actor2 +"')")
  if actor3 != "UNK":
    literals.append("actor('"+ movie_title +"','"+ actor3 +"')")             ####  found the actors for queries 7 , 8 and 9

  language = clean_text(getattr(row, 'language'))
  if language != "UNK":
    literals.append("language('"+ movie_title +"','"+ language +"')")      #### found the language for query 10

  prod_companies = getattr(row, 'production_companies')
  prod_companies = ast.literal_eval(prod_companies)                #### removing double quotes i.e turning "[]" into []
  for studio in prod_companies: 
    studio_id = str(studio['id'])
    literals.append("prod_studio('"+ movie_title +"','"+ studio_id +"')")        #### found production studios for query 12
    
  prod_countries = getattr(row, 'production_countries')
  prod_countries = ast.literal_eval(prod_countries)                #### removing double quotes i.e turning "[]" into []
  for country in prod_countries: 
    country_iso = country['iso_3166_1']
    literals.append("prod_country('"+ movie_title +"','"+ country_iso +"')")        #### found production countries for query 13
  
  # title_year is the same as the year in the date, it has been tested
  year = getattr(row, 'title_year')
  if year != "UNK":
    year = str(int(year))     #### turn to int cause it appeared as 2019.0 instead of 2019 then turn it back to str for concat
    literals.append("year('"+ movie_title +"','"+ year +"')")               #### found the title year for query 14
    year = int(year)
    decade =  year - (year % 10)
    decade = str(decade)   
    literals.append("decade('"+ movie_title +"','"+ decade +"')")
  ### YOUR CODE HERE /\

#Η Prolog θέλει τα κατηγορήματά της με την σειρά 
literals.sort()
for literal in literals:
  prolog.assertz(literal)
  #### printing for the movie "Avatar" for testing purposes
  if literal.startswith("genre('Avatar"):
    print(literal)
  if literal.startswith("director('Avatar"):
    print(literal)
  if literal.startswith("plot_keyword('Avatar"):
    print(literal)
  if literal.startswith("actor('Avatar"):
    print(literal)
  if literal.startswith("language('Avatar"):
    print(literal)
  if literal.startswith("prod_studio('Avatar"):
    print(literal)
  if literal.startswith("prod_country('Avatar"):
    print(literal)
  if literal.startswith("year('Avatar"):
    print(literal)
  if literal.startswith("decade('Avatar"):
    print(literal)
  if literal.startswith("color('Avatar"):
    print(literal)
      
  #print (literal +'.')

#Επίσης μπορούμε να κάνουμε consult ένα έτοιμο αρχείο στον κόσμο όπως παρακάτω
#prolog.consult(path + "db.pl")
````

````python
def simple_recommender_1(movie):
  """
  This function runs find_sim_1 and returns a set of the similar movies provided by this rule.
  """
  s = set()
  q = prolog.query("find_sim_1('" + movie +"',M)")
  for soln in q:
    m = soln['M'] 
    if m not in s:
      s.add(soln['M'])
  q.close()
  answers = s
  return answers

def simple_recommender_2(movie):
  """
  This function runs find_sim_2 and returns a set of the similar movies provided by this rule.
  """
  s = set()
  q = prolog.query("find_sim_2('" + movie +"',M)")
  for soln in q:
    m = soln['M'] 
    if m not in s:
      s.add(soln['M'])
  q.close()
  answers = s
  return answers

def simple_recommender_3(movie):
  """
  This function runs find_sim_3 and returns a set of the similar movies provided by this rule.
  """
  s = set()
  q = prolog.query("find_sim_3('" + movie +"',M)")
  for soln in q:
    m = soln['M'] 
    if m not in s:
      s.add(soln['M'])
  q.close()
  answers = s
  return answers

def simple_recommender_4(movie):
  """
  This function runs find_sim_4 and returns a set of the similar movies provided by this rule.
  """
  s = set()
  q = prolog.query("find_sim_4('" + movie +"',M)")
  for soln in q:
    m = soln['M'] 
    if m not in s:
      s.add(soln['M'])
  q.close()
  answers = s
  return answers

def simple_recommender_5(movie):
  """
  This function runs find_sim_5 and returns a set of the similar movies provided by this rule.
  """
  s = set()
  q = prolog.query("find_sim_5('" + movie +"',M)")
  for soln in q:
    m = soln['M'] 
    if m not in s:
      s.add(soln['M'])
  q.close()
  answers = s
  return answers

def our_simple_recommender(movie):
  """ 
  This function runs all the find_sim in reverse i.e from find_sim_5 to find_sim_1 and returns a 
  sorted list of the similar movies in descending order i.e from most similar (find_sim_5) to less similar (find_sim_1) .
  """ 
  l = []   #### we use a list so we can append movies to it's tail , whereas with sets it's not so simple since it is hashed
  for i in range(5,0,-1): ### find the most similar ones first i.e we start from find_sim_5 and go down
      str_i = str(i)
      q = prolog.query("find_sim_" + str_i + "('"+ movie +"',M)")
      for soln in q:
          m = soln['M'] 
          if m not in l:
            l.append(soln['M'])
      q.close()
  return l  
````

Πέρα από τον recommender που ψάχνει ομοιότητα με απλούς κανόνες Prolog , δημιουργήσαμε και έναν recommender που χρησιμοποιεί ένα σύστημα πόντων για να βρει ομοιότητα μεταξύ των ταινιών:

````python
""" 
We have created 14 queries in part 1.b . We will assign each query points based on how important we think it is.
In our opinion, the genre is the most important factor , with the director , the actors and the plot following up.
The other factors are not so important.
Query 1 : >3 genres the same  ------------> 5 points 
Query 2 : 3 genres the same   ------------> 3 points
Query 3 : 1 genre the same    ------------> 1 point
Query 4 : same director       ------------> 5 points 
Query 5 : >3 plot keywords the same ------> 4 points 
Query 6 :  1 < keywords < 3   ------------> 2 points 
Query 7 : all 3 actors the same  ---------> 5 points 
Query 8 : 2 actors the same   ------------> 3 points 
Query 9 : 1 actor the same    ------------> 1 point 
Query 10 : same language      ------------> 1 point
Query 11 : same color         ------------> 3 points , not so important for colored films but important for black and white
Query 12 : production studio  ------------> 2 points
Query 13 : production country ------------> 2 points
Query 14 : same decade        ------------> 1 point 
"""
def point_recommender(movie):
    dict = {}
    q1 = prolog.query("similar_genre('" + movie +"',M)")
    for soln in q1:
        m = soln['M'] 
        if m not in dict:
          dict[m] = 0
        dict[m] += 5 ### add 5 points for query 1
    q1.close()
    q2 = prolog.query("mostly_similar_genre('" + movie +"',M)")
    for soln in q2:
        m = soln['M'] 
        if m not in dict:
          dict[m] = 0
        dict[m] += 3 ### add 3 points for query 2
    q2.close()
    q3 = prolog.query("less_similar_genre('" + movie +"',M)")
    for soln in q3:
        m = soln['M'] 
        if m not in dict:
          dict[m] = 0
        dict[m] += 1 ### add 1 point for query 3
    q3.close()
    q4 = prolog.query("same_director('" + movie +"',M)")
    for soln in q4:
        m = soln['M'] 
        if m not in dict:
          dict[m] = 0
        dict[m] += 5 ### add 5 points for query 4
    q4.close()
    q5 = prolog.query("same_plot('" + movie +"',M)")
    for soln in q5:
        m = soln['M'] 
        if m not in dict:
          dict[m] = 0
        dict[m] += 4 ### add 4 points for query 5
    q5.close()
    q6 = prolog.query("similar_plot('" + movie +"',M)")
    for soln in q6:
        m = soln['M'] 
        if m not in dict:
          dict[m] = 0
        dict[m] += 2 ### add 2 points for query 6
    q6.close()
    q7 = prolog.query("three_actors('" + movie +"',M)")
    for soln in q7:
        m = soln['M'] 
        if m not in dict:
          dict[m] = 0
        dict[m] += 5 ### add 5 points for query 7
    q7.close()
    q8 = prolog.query("two_actors('" + movie +"',M)")
    for soln in q8:
        m = soln['M'] 
        if m not in dict:
          dict[m] = 0
        dict[m] += 3 ### add 3 points for query 8
    q8.close()
    q9 = prolog.query("one_actor('" + movie +"',M)")
    for soln in q9:
        m = soln['M'] 
        if m not in dict:
          dict[m] = 0
        dict[m] += 1 ### add 1 points for query 9
    q9.close()
    q10 = prolog.query("same_language('" + movie +"',M)")
    for soln in q10:
        m = soln['M'] 
        if m not in dict:
          dict[m] = 0
        dict[m] += 1 ### add 1 points for query 10
    q10.close()
    q11 = prolog.query("same_color('" + movie +"',M)")
    for soln in q11:
        m = soln['M'] 
        if m not in dict:
          dict[m] = 0
        dict[m] += 3 ### add 3 points for query 11
    q11.close()
    q12 = prolog.query("common_studio('" + movie +"',M)")
    for soln in q12:
        m = soln['M'] 
        if m not in dict:
          dict[m] = 0
        dict[m] += 2 ### add 2 points for query 12
    q12.close()
    q13 = prolog.query("common_country('" + movie +"',M)")
    for soln in q13:
        m = soln['M'] 
        if m not in dict:
          dict[m] = 0
        dict[m] += 2 ### add 2 points for query 13
    q13.close()
    q14 = prolog.query("same_decade('" + movie +"',M)")
    for soln in q14:
        m = soln['M'] 
        if m not in dict:
          dict[m] = 0
        dict[m] += 1 ### add 1 point for query 14
    q14.close()
    answers = list(sorted(dict.items(), key=lambda item: item[1]))
    answers.reverse()
    return answers
````

Θα εκπαιδεύσουμε τον recommender μας για 3 , 7 , 10 , 20 , 30 , -1(δηλαδή όλες) ταινίες και θα σημειώσουμε τα αποτελέσματα των πειραμάτων μας στο τέλος του Μέρους 3. Συγκεκριμένα, θα σημείωσουμε τα overall αποτελέσματα (δηλαδή μέσοι όροι απ'όλες τις εκπαιδεύσεις) και συμπεράσματα πάνω στην χρησιμότητα του recommender.

````python
from tqdm.notebook import tqdm
from sklearn.metrics import precision_score, recall_score, f1_score
import numpy as np
import random

rating_weights_given = {0: -1, 1: -0.5, 2:0, 3:0, 4:0.5, 5:1}
rating_weights_heavy = {0: -3, 1: -1, 2:0, 3:1, 4:2, 5:5}
rating_weights_middle = {0: -1, 1: -0.5, 2:0, 3:0.5, 4:1, 5:2}

rating_weights_very_heavy = {0: -10, 1: -5, 2:-2, 3:0, 4:5, 5:10}

score_weights_given = {i:1 for i in range(5)}
score_weights_linear = {i:i + 1 for i in range(5)} 
score_weights_exponential = {0: 0.5 , 1 : 1 , 2 : 2 , 3 : 4 , 4 : 8 }

def train_recommender(ratings, rating_weights, score_weights, number_of_movies = 10):
    """
    Στην συνάρτηση αυτή μπορούμε να ορίζουμε ποιο υποσύνολο των αξιολογήσεων θα χρησιμοποιήσουμε για το train μαζί με τα βάρη ομοιότητας και σκορ
    Σε συνδυασμό με τον αριθμό των ταινιών που θέλουμε να χρησιμοποιήσουμε σαν σύνολο δεδομένων π.χ. 10 από τις 100 ή 3 από τις 100 κ.ο.κ
    Αν θέλουμε να χρησιμοποιήσουμε όλες τις ταινίες σαν training set τότε ορίζουμε το number_of_movies = - 1
    """

    if number_of_movies > len(ratings):
        number_of_movies = len(ratings)


    if number_of_movies != -1:
        indexes = random.sample(range(len(ratings)), number_of_movies)
        ratings = ratings.iloc[indexes]

    movie_score = {}
    for row in tqdm(ratings.itertuples(index=True, name='Pandas')):
        movie = clean_text(getattr(row, 'movie_title'))
        rating = getattr(row, 'rating')

        similar_movies_1 = simple_recommender_1(movie)
        similar_movies_2 = simple_recommender_2(movie)
        similar_movies_3 = simple_recommender_3(movie)
        similar_movies_4 = simple_recommender_4(movie)
        similar_movies_5 = simple_recommender_5(movie)

        for similar_movie in similar_movies_1:
            if similar_movie not in movie_score:
                movie_score[similar_movie] = rating_weights[int(rating)] * score_weights[0]
            else:
                movie_score[similar_movie] += rating_weights[int(rating)] * score_weights[0] # το weight θα το ορίσετε ανα επίπεδο ομοιότητας οι πολύ όμοιες ταινίες θα έχουν μεγαλύτερο βάρος
    
        for similar_movie in similar_movies_2:
            if similar_movie not in movie_score:
                movie_score[similar_movie] = rating_weights[int(rating)] * score_weights[1]
            else:
                movie_score[similar_movie] += rating_weights[int(rating)] * score_weights[1]

        for similar_movie in similar_movies_3:
            if similar_movie not in movie_score:
                movie_score[similar_movie] = rating_weights[int(rating)] * score_weights[2]
            else:
                movie_score[similar_movie] += rating_weights[int(rating)] * score_weights[2]   

        for similar_movie in similar_movies_4:
            if similar_movie not in movie_score:
                movie_score[similar_movie] = rating_weights[int(rating)] * score_weights[3]
            else:
                movie_score[similar_movie] += rating_weights[int(rating)] * score_weights[3]

        for similar_movie in similar_movies_5:
            if similar_movie not in movie_score:
                movie_score[similar_movie] = rating_weights[int(rating)] * score_weights[4]
            else:
                movie_score[similar_movie] += rating_weights[int(rating)] * score_weights[4]

    return movie_score


# Αυτό είναι ένα παράδειγμα για το πως θα μπορούσε να υλοποιήθει η predict.
# Έχουμε ορίσει ότι μια ταινία θα έπρεπε να είναι προτεινόμενη αν είχε σκορ > 0.
def predict_example(ratings, movie_score):
    real, pred = [], []
    for i, row in enumerate(ratings.itertuples(index=True, name='Pandas')):
        movie = clean_text(getattr(row, 'movie_title'))
        rating = getattr(row, 'rating')

        if movie in movie_score: #αν έχουμε σχηματίσει βαθμολογία για την ταινία αυτή
            pred.append(int(movie_score[movie] > 0)) #heuristic για το αν μια ταινία είναι προτεινόμενη
            real.append(int(rating > 3))# έτσι ορίζουμε ότι μια ταινία θα έπρεπε να είναι προτεινόμενη
            #η συνθήκη αυτή δεν μπορεί να αλλάξει
        else: #δεν μπορούμε να προτείνουμε κάτι για το όποιο δεν έχουμε σχηματίσει εικόνα
            pred.append(0)
            real.append(int(rating > 3))

    return real, pred


def get_metrics(real, pred):
    metrics = {}
    metrics["precision"] = precision_score(real, pred, zero_division = 0) ## added zero_division to suppress warings
    metrics["recall"] = recall_score(real, pred)
    metrics["f1"] = f1_score(real, pred)
    return metrics
````
